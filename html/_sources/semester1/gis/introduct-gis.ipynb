# Intro to GIS

Năm 1960 GIS hoạt động đầu tiên tại Ottawa, Canada được gọi là hệ thống thông tin địa lí Cannada được sử dụng để lưu trữ, phân tích, thao tác dữ liệu, thu thập đất đai cho nông thôn.
Cho đến năm 1980-9: GIS bắt đầu được thương mại hóa và đến ngày nay thời kì cởi mở và sử dụng phổ biến.

Tôi hiểu GIS như là bản đồ địa lí được không?

Các dạng tồn tại của đối tượng trong thế giới thực:
|Ngữ nghĩa | Không Gian | Thời gian| Ví dụ|
|:---------|:---------:|:---------|:---------|
|y|n|n||
|n|y|n|Khối 3D chụp từ vệ tinh|
|n|n|y|thứ, năm|
|y|y|n|con sông|

Ví dụ : Cây cầu
* Không gian: Tập hợp các đường: điểm đầu, điểm cuối, các điểm trung gian, 1 mặt tập hợp các đỉnh đa giác, tọa độ cây cầu trên bản đồ
* Thời gian: Ngày khởi công, ngày hoàn thành, ngày sửa chữa
* Ngữ nghĩa: Tên cây cầu, loại ccaay cầu, chủ đầu tư, tổng kinh phí.......

Chiều thời gian không chỉ cần thiết trong csdl địa lý mà một httt cũng cần thiết

Ví dụ: NV( MSNV, MSCV, ngaynhanchuc)

- Điểm thời gian(instant): một điểm trên trục thời gian
- Đoạn thời gian
- Khoảng thời gian
- Độ phân giải và thứ tự của thời gian
- Độ phân giải(grabularity): là đoạn thời gian nhỏ nhất trên trục thời gian. Thời gian liên tục thì trục thời gian đẳng cấu với tập số thực. Thời gian rời rạc thì trục thời gian bao gồm những điểm xác định.

GIS hình thành từ khoa học liên ngành: KHMT, Toán học, địa lý học, bản đồ học...

GIS - Geographic Information System





1. Hệ thống thông tin địa lí
- Hệ thống là tập hợp các thành phần con phối hợp với nhau một cách khoa học để đạt được 1 mục tiêu

- Thông tin: Data --> thông tin --> tri thức --> wistom

    Hệ thống thông tin:
    * Lưu trữ
    * Xử lí
    * Biểu diễn
    * Truyền
    * Tiếp nhận

- Yếu tố không gian: (x, y), (kinh độ, vĩ độ)

=> Hệ thống thông tin địa lý là sự kế hợp của hệ thống thông tin và yếu tố không gian

- Đối tượng trong HTTTDL bao gồm 3 thuộc tính: Ngữ nghĩa, Không gian, Thời gian

- Tip của thầy: 

    * Bài toán lớn là tập hợp các bài toán nhỏ bị che dấu, vì vậy các em cần xác định được những bài toán nhỏ
    * Viết nhật ký hằng ngày theo dõi sự phát triển của bản thân
    * Cách tốt nhất để chuẩn bị câu trả lời cho các câu hỏi khi bảo vệ luận văn là đi nghe các buổi bảo vệ và ghi lại những câu hỏi đó, hoặc xin các câu hỏi từ thầy cô hướng dẫn.
2. Mối quan hệ không gian
- Thứ tự
- Topology
- Độ đo: 1D - Chiều dài, 2D - Chu vi, diện tích, 3D - Thể tích
3. Biểu diễn một đối tượng không gian
- Vector: Tọa đổ các điểm trên đường viền, thường dùng để biểu diễn các đối tượng nhân tạo
- Raster: Chia đối tượng thành các pixel, thường dùng để biểu diễn các đối tượng tự nhiên

4. Các chức năng của GIS
- Thu nhận dữ liệu
- Cấu trúc
- Thao tác
- Phân tích
- Trình bày dữ liệu

5.Kết luận
* GIS là ngành khoa học được xây dựng và phát triển trên nền tảng của một số ngành khoa học khác.
* Các GIS có thể tồn tại dưới một số mô hình, các mô hình càng mịn càng mô tả rõ hệ thống.
* GIS là HTTT có tính trực quan cao nhờ yếu tố không gian.
* GIS còn là HTTT theo thời gian nhằm biểu diễn lịch sử biến đống các đối tượng trong vòng đời của chúng.
* Các ứng dụng GIS là rất phổ biến, đa dạng
* Đầu tư GIS là khá tốn kém do đòi hỏi CS hạ tầng là các công nghệ cao.
* GIS thành công phụ thuộc: khảo sát, phân tích, thiết kế, xây dựng, kiểm định, khai thác, bảo trì, cập nhật.
